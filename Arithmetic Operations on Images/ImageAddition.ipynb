```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image1 = cv2.imread(r'cosmic1.jpg')
plt.imshow(image1)
plt.show()
```

```python
image2 = cv2.imread(r'nature.jpg')
plt.imshow(image2)
plt.show()
```

```python
weightedSum = cv2.addWeighted(image1,0.5,image2,0.4,0)
```

```python
cv2.imshow('Summed Image',weightedSum)
plt.imshow(weightedSum)
plt.show()
```
