#Import Modules

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

#Load Image 

```python
#image is loaded using cv2.imread() method,here flag is zero ,because image is loaded in grayscale mode.
'''
Syntax:
  cv2.imread(path,flag)
Parameters:
  path: string representing the path of the image to be read.
  flag: specifies the way in which image should be read.
'''
img=cv2.imread("clahe_input1.png",0)
cv2_imshow(img)
```

#Apply CLAHE operation


```python
#At first,CLAHE is created and then applied on the image.
'''
Syntax:
  cv2.createCLAHE(clipLimit=40,tileGridSize=(8,8))
Parameters:
  clipLimit sets the threshold for contrast limiting. The default value is 40.
  tileGridSize sets the number of tiles in the row and column. By default this is 88.
'''
clahe=cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
CLAHEed=clahe.apply(img)
```

#Display the output image

```python
cv2_imshow(CLAHEed)
```
